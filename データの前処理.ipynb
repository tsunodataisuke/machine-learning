### iloc：行列を番号で指定する
行列を番号で指定します。第一引数で行を、第2引数で列を指定します。引数には数値やスライス、リストを指定できます
#### [  ,   ] この「,」が第一引数と第二引数の境界線になっており、[:3,0]は第一引数は:3まで。第二引数が0である。

### 文字列に限らずリストなどのシーケンスでは、部分的に要素を抜き出すことができます。この操作を スライス と呼んでいます

### 文字列変数名[開始位置:終了位置] ただしこれは[ : ]なのでスライスだ。スライスは１から３まで抜き出すのように範囲指定する。リストは[,,,]で指定して指定したものを抜き出す
### と記述します。こうすることで、開始位置の場所から <span style="color: red; ">終了位置 - 1</span> の場所までの文字列を取得できます。終了位置には気をつけましょう。

```python
# 行をスライス、列をリストで指定
dataset.iloc[1:3, [0, 2, 4]]  #行を１番目開始から３番目(-1番目)終了まで取得、列を０番目、２番目、４番目を取得。
                              #[開始:終了:増分]← 「:」を使う 　      　[0,2,4] 　　　「,」を使う　のは、０番目と２番目と４番目を指定して取得する。  混同するな
```

```python
# 行列ともリストで指定
dataset.iloc[:5, 3:6]  #行は０番目から５(-1)番目まで、列は３番目から６(-1)番目まで
```

```python
# 列の指定を省略すると全ての列となる
dataset.iloc[:3] #列の指定がない点に注目
```

```python
# 全ての行を選択するには「:」を指定する
dataset.iloc[:,0] #行を全て取得する、列は０番目
```

### 正規化
## <span style="color: red; ">それぞれの説明変数で値の平均や標準偏差に差がありすぎると、値の大きい説明変数に引っ張られて学習に悪影響を及ぼす場合があります。そのような場合に有効なのが正規化です。</span>
正規化はデータを一定の範囲内（-1〜1、または0〜1など）に収まるように加工します。データの正規化には2種類あります。

```python
# 正規化用にdatasetの一部を抜き出し
dataset4 = dataset[["推定年俸", "打点"]]
dataset4.head()
```

#### z-score normalization
**平均値0、標準偏差1に変換します。**標準化（Standardization）とも呼ばれます。正規分布にしたがっているデータは主にこちらを使用します。計算式は「（値 - 平均値）／標準偏差」となります。標準化を行うにはscikit-learnサイキットラーンの StandardScaler を使用します。

```python
# z-score normalization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
dataset4_std = sc.fit_transform(dataset4)
dataset4_std
```

### Min-Max Normalization
**最小値0、最大値1に変換します。**最小値と最大値が決まっていて、データが一様分布の場合はこちらを使用します。計算式は「（値 - 最小値）／（最大値 - 最小値）」となります。正規化を行うにはscikit-learnの MinMaxScaler を使用します。

```python
# Min-Max Normalization
from sklearn.preprocessing import MinMaxScaler
ms = MinMaxScaler()
dataset4_norm = ms.fit_transform(dataset4)
dataset4_norm
```

## 欠損値の扱い

機械学習を行うために収集したデータには、数値データのはずなのに空白が入っていたり、時系列データの一部が抜けていたりと、一部のデータが欠けていることがよくあります。このように欠けているデータのことを「欠損値」と呼びます。欠損値が存在すると機械学習の途中でエラーになる、学習が正しく行えないなどの悪影響があります。

欠損値の扱いは欠損が発生した状況などにより異なります。基本的に学習の精度に影響が出ないよう、0や平均値、中央値などの値を設定しますが、1行のうち大部分のデータが欠けてしまっている場合には学習対象外として欠損値を含む行をまるごと除去するなどの対応を取ります。

どちらの場合でも、データを加工することになるため、対応の理由を説明できるようにしておく必要があります。


## 欠損値の確認
列ごとの欠損値の有無を確認するには isnull 関数を使用します。sumを付けることで欠損値の行数を確認できます。 describe() 関数や info() 関数を使っても同様の確認が行なえます

```python
# 欠損値の有無の確認
dataset.isnull().sum()
```

##### 【関数名】　　　　【説明】
##### dropna　　　→欠損値を含む行の削除
##### fillna　　　　→欠損値の補完

```python
# describeのcountでも確認できる
dataset.describe().count()
```

```python
# infoでも確認できる
dataset.info()
```

## 欠損値の補完

欠損値をなんらかの値で補完するには fillna 関数を使用します。例として「年数」の欠損値に平均値を代入してみましょう。meanは平均値を求める関数です。

```python
# 欠損値に平均値を代入。「dataset['年数'].mean()」で平均値を求めている
dataset['年数'] = dataset['年数'].fillna(dataset['年数'].mean())
dataset.head()
```

## [再掲]次元削減と主成分分析

説明変数の中には「打率と出塁率」など、相関性の高いものが複数含まれています。それらをそのまま機械学習にかけると以下のようなデメリットがあります。

#### ・それらの説明変数が、他の説明変数に比べてモデルの学習に強く影響を及ぼすため、過度に学習されたり、モデルの性能が向上しないなどの悪影響となる可能性がある
#### ・まとめれば1つで済む説明変数を複数で計算させるので、その分学習時間が長くなる
#### ・4次元以上になるとグラフによる可視化が行いづらい
#### これを解消するのが次元削減です。次元削減とは次元（列）を削減するということですが、主に元の情報量を保ったまま相関性の高い複数の説明変数をまとめる方法という意味で使われます。

## 主成分分析(PCA)による次元削減

次元削減の代表的な手法が主成分分析（principal component analysis, PCA)です。主成分分析は列（成分）に重みを付けて1つの列にまとめる手法です。詳しい考え方までは理解しなくても大丈夫ですが、レッスンの補足に掲載していますので、必要に応じて参考にしてください。主成分分析はscikit learnの decomposition.PCA クラスを使って行います。

```python
from sklearn.decomposition import PCA
```

```python
# batting_avgとon_base_perを取り出して主成分分析
pca = PCA(n_components=1).fit(dataset2[['batting_avg', 'on_base_per']])
```

```python
# 主成分分析した結果を基にデータ変換を行い成分を取得する
values = pca.transform(dataset2[['batting_avg', 'on_base_per']])
```

```python
# 得られた成分をdataset2に追加
dataset2['PCA_val'] = values[:,0]
```

```python
dataset2.head()
```

得られた成分と推定年俸との関係性をグラフ化してみましょう。

```python
# 推定年俸：PCA成分
dataset2.plot(kind='scatter', x='salary', y='PCA_val')
```

主成分分析前の2つのグラフに似た形状となっていますね。もとの成分の関係性を保ったまま、1つの成分にまとめられていることが確認できます。
